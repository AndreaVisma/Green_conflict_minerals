## Price data available from the World Bank

Not sure if there's other sources that make sense, so far I couldn't really find anything tooo useful

```python
import pandas as pd
import matplotlib.pyplot as plt

import os

%matplotlib inline
```

```python
THIS_FOLDER = os.getcwd()

data_path = os.path.join(THIS_FOLDER, 'Data_sources/price_data.xlsx')

df = pd.read_excel(data_path, sheet_name = "Monthly Prices", usecols = "A, BK:BT")
df['Year'] = df['Year'].apply(lambda x: x[:4] + " " + x[5:])
df['Year'] = pd.to_datetime(df['Year'])
df.set_index("Year", inplace = True)

print("datafrme dimensions: ", df.shape)
print("     ")
print(df.head())
```

```python
green_minerals = ["ALUMINUM", "COPPER", "NICKEL", "SILVER"]

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex = True, figsize=(9, 9))


ax1.plot(df.index, df[green_minerals[0]], color = 'blue')
ax1.set_title(green_minerals[0])
ax1.set_ylabel('$/mt')
ax2.plot(df.index, df[green_minerals[1]], color = 'green')
ax2.set_title(green_minerals[1])
#ax2.set_ylabel('$/mt')
ax3.plot(df.index, df[green_minerals[2]], color = 'orange')
ax3.set_title(green_minerals[2])
ax3.set_ylabel('$/mt')
ax3.tick_params(axis='x', rotation=45)
ax4.plot(df.index, df[green_minerals[3]], color = 'red')
ax4.set_title(green_minerals[3])
ax4.set_ylabel('$/troy oz')
ax4.tick_params(axis='x', rotation=45)

fig.savefig('long-range-prices.pdf')
```
